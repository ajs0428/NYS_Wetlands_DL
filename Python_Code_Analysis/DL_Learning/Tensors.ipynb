### Creating torch tensors

Create a 3-dimensional tensor with 12 rows and 12 cols

```python
t4 = torch.zeros(size = (3, 12, 12))
print(t4)
```

Create 2 tensors with 4 rows and cols with 1s 

```python
t5 = torch.ones(size=(2,4,4))
print(t5)
```

Create a 1 dimension tensor with a range of values starting at 10 and ending before 100 by 5

```python
t6 = torch.arange(start=10, end=100, step=5)
print(t6)
```

### Reshaping Tensors

Create 1-d tensor of 32 ones

```python
arr2 = np.ones(32)
t17 = torch.tensor(arr2)
print(t17)
```

reshape to 2-d array with 4 rows and 4 cols

```python
t18 = torch.reshape(t17, (2, 4, 4))
print(t18)
```

Use `-1` if the dimension is unknown 

```python
t19 = torch.reshape(t17, (-1, 4, 4))
t19
```

```python
t20 = torch.reshape(t17, (4, 4, -1))
print(t20)
```

`torch.flatten` will turn any tensor into a 1-d tensor

```python
t21 = torch.flatten(t20)
print(t21)
```

`view()` method can show tensor with a particular dimension

```python
t22 = t17.view((2,16))
print(t22)
```

### Indexing tensors

- Indexing tensors starts at 0
- `:` is used for ranges
- last index not included
- a `colon` by itself selects all elements in that dimension

```python
t23 = torch.randint(low=0, high=256, size = (8,8))
print(t23)
```

select all the rows, and 0-4 of the columns

```python
t24 = t23[:,0:4]
print(t24)
```

select 0-4 of the rows and all the cols

```python
t25 = t23[0:4,:]
print(t25)
```
